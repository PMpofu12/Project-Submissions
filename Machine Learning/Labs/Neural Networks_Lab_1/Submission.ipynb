```python
# (d) Write a function which propagates a batch of data through a neural network.

# The following function will be used for the sigmoid activation function. 
def sigmoid(x):
    sig = 1/(1+np.exp(-x))
    return sig

# The following function will be used to obtain the inputs for the propagation function.
def getInputs(data,weights,bias):
    inputs = np.dot(data,weights) + bias
    return inputs

def propagate(dimensions,data):
    outputLayer = ([]) #contains output for every layer
    inputLayer= ([]) #contains input for every layer
    
    for x in range(len(neural_network)):
        layer = neural_network[x] # Access the current network
        weight_matrix = layer[0] # Access the weight matrix at the current layer.
        bias_layer = layer[1] # Access the bias layer at the current layer.
    
        if x == 0: # Uses the training data as input for the first iteration of the loop
            outputLayer = getInputs(data,weight_matrix,bias_layer)
            inputLayer = outputLayer # Output becomes input for next layer
        else:
            outputLayer = sigmoid(inputLayer)
            outputLayer = getInputs(outputLayer,weight_matrix,bias_layer)
            inputLayer = outputLayer # Output becomes input for next layer  
    outputLayer=sigmoid(inputLayer)        
    return outputLayer 
```
